```python
import numpy as np
import pandas as pd
import os
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
csv = pd.read_csv('./result.csv', sep=',', header=None)
csv.columns = ["epoch", "loss_rpn_cls", "loss_rpn_regr", "loss_class_cls", "loss_class_regr", "class_acc", "curr_loss"]
csv
```

```python
fig = plt.figure(figsize=(8, 6))
epoch = range(len(csv))

plt.xlabel('epoch')
plt.ylabel('class_acc')
plt.legend('class_acc')

plt.plot(epoch, csv['class_acc'] ,"r",linewidth=1.5, linestyle="-")
```

```python
fig = plt.figure(figsize=(8, 6))
epoch = range(len(csv))

plt.xlabel('epoch')
plt.ylabel('curr_loss')
plt.legend('curr_loss')

plt.plot(epoch, csv['curr_loss'] ,"r",linewidth=1.5, linestyle="-")
```
